```python
# 본점의 part_nm별 총 판매액을 구해라.
# 1. 본점인 애들만 추출한다.
sh2 = df1[df1['str_nm']=='본점']
sh2
```

```python
# 2. part_nm 가져오기
array = sh2['part_nm'].unique()
array
```

```python
sh2[sh2['part_nm']=='케주얼,구두,아동']['tot_amt'].sum()
```

```python
# 3. 각 카테고리별 총합 구하기
for i in array:
    result = sh2[sh2['part_nm']== i]['tot_amt'].sum()
    print('{}: {}원'.format(i,result))
```

```python
# 간단하게 하는 방법 "그룹핑"
# 본점인 데이터 추출.prt_nm별로그룹핑.tot_amt에대하여.총합.표로보여라
sh2.groupby('part_nm').tot_amt.sum().reset_index(name="총판매액").rename(columns = {'part_nm':'파트'})
```
